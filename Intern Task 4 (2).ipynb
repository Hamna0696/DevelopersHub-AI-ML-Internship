```python
import os
import requests
from getpass import getpass

GROQ_API_KEY = getpass("Groq API Key enter karo: ")

API_URL = "https://api.groq.com/openai/v1/chat/completions"
headers = {
    "Authorization": f"Bearer {GROQ_API_KEY}",
    "Content-Type": "application/json"
}
print("Setup completed successfully!")
```

```python
system_prompt = """
You are a friendly, empathetic, and helpful medical assistant.
Your goal is to provide clear, general educational information about health queries.

CRITICAL SAFETY RULES:
1. You are NOT a doctor. Never provide definitive medical diagnoses or specific prescription dosages.
2. Always include a short disclaimer advising the user to consult a professional healthcare provider for medical concerns.
3. If a query implies a life-threatening emergency (e.g., severe chest pain, difficulty breathing), immediately instruct the user to call emergency services.
"""
print("System prompt and safety guidelines successfully defined!")
```

```python
def get_chatbot_response(user_query):
    emergency_keywords = ["chest pain", "heart attack", "stroke", "severe bleeding", "unconscious"]
    if any(keyword in user_query.lower() for keyword in emergency_keywords):
        return "EMERGENCY WARNING: Please call emergency services immediately (e.g., 1122)!"

    payload = {
        "model": "llama-3.3-70b-versatile",
        "messages": [
            {
                "role": "system",
                "content": system_prompt
            },
            {
                "role": "user",
                "content": user_query
            }
        ],
        "max_tokens": 1000,
        "temperature": 0.7
    }

    response = requests.post(API_URL, headers=headers, json=payload, timeout=30)
    if response.status_code == 200:
        return response.json()["choices"][0]["message"]["content"].strip()
    else:
        return f"Error: {response.status_code} - {response.text}"

print("Chatbot function ready!")
```

```python
print("Medical Chatbot Ready! Type 'quit' to exit.\n")
while True:
    user_input=input("You:")
    if user_input.lower() in ["quit" , "exit","bye"]:
        print("Chatbot:Take care! Goodbye!")
        break
    if not user_input.strip():
        continue
    response=get_chatbot_response(user_input)
    print(f"Chatbot:{response}\n")
```
