```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('iris')
print("Dataset Sucessfully Loaded")
```

```python
print("Dataset Shape")
print(df.shape)
print("\n")

print("Columns Names")
print(df.columns.tolist())
print("\n")

print("First 5 Rows")
df.head()
```

```python
print("Data Info")
df.info()
print("\n")

print("Statistical Summary")
df.describe()
```

```python
#VISUALIZATION
sns.set_theme(style="whitegrid")

#SCATTER PLOT:
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='sepal_length',y='sepal_width',hue='species',palette='Set1')
plt.title('Scatter Plot:Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length(cm)')
plt.ylabel('Sepal Width(cm)')
plt.show()
```

```python
#HISTOGRAM 
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='petal_length',hue='species',kde=True,bins=20,multiple='stack',palette='Dark2')
plt.title('Histogram:Distribution of Petal Length')
plt.xlabel('Petal Length(cm)')
plt.ylabel('Count')
plt.show()
```

```python
#BOX PLOT
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x='species',y='petal_width',palette='Pastel1')
plt.title('Box Plot:Petal Width Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width(cm)')
plt.show()
```
